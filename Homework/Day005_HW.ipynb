```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
content = json.loads(r.text)
print('ä¸€æ¬¡å›å‚³ {} ç­†è³‡æ–™'.format(len(content)))
print('è³‡æ–™æ¬„ä½åŒ…å«: \n{}'.format(', '.join(content[0].keys())))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
import pandas as pd

col = ['title', 'createdAt', 'commentCount', 'likeCount']
df = pd.DataFrame(columns=col)

for article in content:
    df = df.append(
        pd.DataFrame([[article[i] for i in col]],
        columns = col)
    )
df
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
import numpy as np

popular_article = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
nonpopular_article = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')

popular_json = json.loads(popular_article.text)
nonpopular_json = json.loads(nonpopular_article.text)

print('ç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸ï¼š {}'.format(np.mean([art['commentCount'] for art in popular_json])))
print('ç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸ï¼š {}'.format(np.mean([art['likeCount'] for art in popular_json])))
print('éç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸ï¼š {}'.format(np.mean([art['commentCount'] for art in nonpopular_json])))
print('éç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸ï¼š {}'.format(np.mean([art['likeCount'] for art in nonpopular_json])))
```
