# Ettoday 網路爬蟲實作練習


* 能夠利用 Request + BeatifulSour 撰寫爬蟲，並存放到合適的資料結構


## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
driver = webdriver.Chrome()
driver.get(url)
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime
import random
from time import sleep

endTime = (datetime.datetime.now() - datetime.timedelta(hours = 2)).strftime('%Y/%m/%d %H:%M')

while driver.find_element_by_class_name("part_list_2").find_elements_by_class_name('date')[-1].text >= endTime:
    sleep(random.randrange(3))
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    
soup = BeautifulSoup(driver.page_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.span.text >= endTime:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import datetime
from selenium.webdriver.support.ui import Select

startTime = (datetime.datetime.now() - datetime.timedelta(days = 3)).strftime('%Y/%m/%d ')+'15:00'
endTime = (datetime.datetime.now() - datetime.timedelta(days = 3)).strftime('%Y/%m/%d ')+'17:00'

yearButtom = Select(driver.find_element_by_name('selY'))
yearButtom.select_by_value(startTime[0:4])

monthButtom = Select(driver.find_element_by_name('selM'))
monthButtom.select_by_value(startTime[5:7])

dayButtom = Select(driver.find_element_by_name('selD'))
dayButtom.select_by_value(startTime[8:10])

driver.find_element_by_xpath('//input[@id="button"]').click()

while driver.find_element_by_class_name("part_list_2").find_elements_by_class_name('date')[-1].text >= startTime:
    sleep(random.randrange(3))
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    
soup = BeautifulSoup(driver.page_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.span.text >= startTime) & (d.span.text < endTime):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
driver.quit()
```
